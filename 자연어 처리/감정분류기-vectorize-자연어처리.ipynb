```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
## 감정분류기
# 긍정/부정 사전을 만들어보세요.
# input()을 이용해서 ,문장을 받으면, 긍정인지 부정인지 판단해보세요.
# if-idf vectorizer를 사용해주세요.
```

```python
 cs_pos= CountVectorizer()
```

```python
cv_neg= CountVectorizer()
```

```python
tfidf_pos=TfidfVectorizer()
```

```python
tfidf_neg=TfidfVectorizer()
```

```python
pos_dict = ["happy", "glad", "good", "well", "thrilled", "pleased","satisfied", "lmpressed",'moved','relaxed','lnspried'] 
neg_dict = ["nervous", "sick", "frustarte", "bored", "sacred", "agitaed", "general", "sore",'tired','worried']
```

```python
emotion =[input('단어를 입력해주세요:')]

cs_pos.fit(pos_dict)
cs_pos.vocabulary_
result1=cs_pos.transform(emotion).toarray()
result1_count =0
for item in result1[0]:
  if(item !=0):
    result1_count = result1_count +1
result1_count

cv_neg.fit(neg_dict)
cv_neg.vocabulary_
result2=cv_neg.transform(emotion).toarray()
result2_count =0
for item in result2[0]:
  if(item !=0):
    result2_count = result2_count +1
result2_count

if result1_count >=1 :
  print('긍정적인단어입니다.')
elif result2_count >=1 :
  print('부정적인단어입니다.')
```

```python
emotion =[input('단어를 입력해주세요:')]

tfidf_pos.fit(pos_dict)
tfidf_pos.vocabulary_
reuslt3 = tfidf_pos.transform(emotion).toarray()
result3_count =0
for item in reuslt3[0]:
  if(item !=0):
    result3_count = result3_count +1
result3_count


tfidf_neg.fit(neg_dict)
tfidf_neg.vocabulary_
reuslt4 = tfidf_neg.transform(emotion).toarray()
result4_count =0
for item in reuslt4[0]:
  if(item !=0):
    result4_count = result4_count +1
result4_count

if result3_count >=1 :
  print('긍정적인단어입니다.')
elif result4_count >=1 :
  print('부정적인단어입니다.')
```
