# 연령대 컬럼을 추가하시오 
컬럼이름: AgeGroup

연령대 : 미성년자, 20대, 30대, 40대, 50대, 60대이상 : 'Minor', '20s', '30s', '40s', '50s', '60+'

```python
ti1=pd.DataFrame(ti,copy=True)
ti1['AgeGroup']=pd.cut(ti.Age,bins=[0,20,30,40,50,60,100], labels=['Minor', '20s', '30s', '40s', '50s', '60+'])
ti1
```

# 각 연령대별 인구 수를 구하시오 

```python
ti1['AgeGroup'].value_counts()
```

```python
ti1['AgeGroup'].value_counts().sort_index() # .sort_index() 인덱스 값을 기준으로 정려
```
