```python
import pandas as pd
import numpy as np
```

```python
ti = pd.read_csv('titanic.csv')
ti
```

```python
ti.info()
```

```python
ti=ti.drop(['Name'],axis=1)
ti
```

```python
ti['Sex']=ti['Sex'].map({'female':1,'male':0}).astype(int) #성별 바꾸기
ti
```

```python
ti=ti.astype({'Age':'int'})
ti
```

```python
ti=ti.astype({'Fare':'int'})
ti
```

```python
ti.info()
```

```python
from sklearn.model_selection import  train_test_split
```

```python
X_data =ti.loc[:,'Pclass':'Fare'] # 문제데이터
y_data =ti.loc[:,'Survived'] #정답데이터
```

```python
# 문제는 대문자 정답은 소문자
X_train, X_test , y_train ,y_test = train_test_split( # 파라미터들
    X_data, # 문제 데이터
    y_data,  # 답데이터
    test_size = 0.2, # 비율
    shuffle =True, # 셔플링여부
    random_state =20 # 시드값
) 
```

```python
X_train, X_test , y_train ,y_test
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors= 11)
```

```python
knn.fit(X_train,y_train)
```

```python
pred  = knn.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
jum =  accuracy_score(y_test,pred) # 원래정답, 내가푼답 

jum # accuracy
```

```python
from sklearn.svm import SVC
```

```python
svc= SVC(kernel='rbf')
```

```python
svc.fit(X_train,y_train)
```

```python
pred2  = svc.predict(X_test)
pred2
```

```python
jum2 =  accuracy_score(y_test,pred2) # 원래정답, 내가푼답 

jum2 # accuracy
```
