```python
import pandas as pd
import numpy as np
```

```python
cf=pd.read_csv('daegu.csv', encoding='utf8')
```

```python
cf
```

```python
cf.info()
```

```python
coffe =(cf.category == '커피숍')
coffe
```

```python
df=cf[coffe]
df
```

```python
df.info()
```

```python
df=df.replace({'close':np.nan},{'close':20221202})
df
```

```python
df= df.astype({'close' :'int'})
df
```

```python
df=df.dropna(axis=0)
df
```

```python
df.info()
```

```python
df.reset_index(drop = False,inplace=True)
df
```

```python
address=[]
for i in df['adress']:
    address.append(i.split(' ')[0])
df['adress'] = address
df
```

```python
df.rename(columns={'adress':'address'}, inplace=True)
df
```

```python
df['open'] = pd.to_datetime(df['open'], format='%Y%m%d')
df['close'] = pd.to_datetime(df['close'], format='%Y%m%d')
df.info()
```

```python
df
```

```python
df['b_day'] = df['close'] - df['open']
df
```

```python
df.drop(['index'], axis=1, inplace=True)
```

```python
df
```

```python
df['franchise'] = df['name'].apply(lambda x : '점' in x )
```

```python
df
```

```python
df['franchise'] = df['franchise'].apply(lambda x: 1 if x==True else 2)
```

```python
df
```

```python
df['area'] = df['area'].str.replace(',','')
```

```python
df['area'] = df['area'].astype(float)
```

```python
df.describe()
```

```python
bins = [0, 28.47, 46.42, 81.78, 1300]
df['size'] = pd.cut(df['area'], bins = bins, labels = [1,2,3,4])
df
```

```python
df['b_day'] = df['b_day'].astype('str')
```

```python
df['b_day'] = df['b_day'].str.split(' ').str[0]
```

```python
df
```

```python
df['addr'] = df['address'].str.replace('남구','1')
df['addr'] = df['addr'].str.replace('달서구','2')
df['addr'] = df['addr'].str.replace('달성군','3')
df['addr'] = df['addr'].str.replace('동구','4')
df['addr'] = df['addr'].str.replace('북구','5')
df['addr'] = df['addr'].str.replace('서구','6')
df['addr'] = df['addr'].str.replace('수성구','7')
df['addr'] = df['addr'].str.replace('중구','8')
df
```

```python
df.drop(['category'], axis=1, inplace=True)
```

```python
df['mub'] = df['mub'].str.replace('Y','1')
df['mub'] = df['mub'].str.replace('N','2')
```

```python
df
```

```python
df['영업상태'] = df['open close'].apply(lambda x: '영업' if x ==1 else "폐업" )
df
```

```python
df.drop((df.loc[df['address'] == '경상북도']).index, axis=0, inplace=True)
```

```python
df.to_csv('coffee.csv', encoding='cp949')
```
