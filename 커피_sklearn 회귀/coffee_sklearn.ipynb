### 트레이닝셋 분리

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(coffee_df, target_df, test_size=0.2, random_state=11)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state=11)
rf_clf = RandomForestClassifier(random_state=11)
lr_clf = LogisticRegression()
knn5 = KNeighborsClassifier(n_neighbors=5)
knn10 = KNeighborsClassifier(n_neighbors=10)
svc_rbf = SVC(kernel='rbf')
```

```python
clf_list = [dt_clf, rf_clf, lr_clf, knn5, knn10, svc_rbf]
```

### 교차검증

```python
from sklearn.model_selection import cross_val_score
```

```python
def callKFold_all(clf_list):
    for clf in clf_list:
        scores = cross_val_score(clf, X_train, y_train, cv=5)
        for iter_count, accuracy in enumerate(scores):
            print("교차 검증 {0} 정확도: {1:.4f}".format(iter_count, accuracy))

        print("평균 정확도: {0:.4f}".format(np.mean(scores)))
```

```python
final_result = callKFold_all(clf_list)
final_result
```

### lr_clf (LogisticRegression)

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters_lr = {'penalty':['l3','l2', 'l1'], 'C':[0.01, 0.1, 1, 1, 5, 10]}
```

```python
grid_lclf = GridSearchCV(lr_clf, param_grid=parameters_lr, scoring='accuracy', cv=5)
grid_lclf.fit(X_train, y_train)
```

```python
best_lclf = grid_lclf.best_estimator_
best_lclf
```

```python
best_params = grid_lclf.best_params_
best_params
```

```python
best_score = grid_lclf.best_score_
best_score
```

```python
pred = best_lclf.predict(X_test)
pred
```

```python
accuracy = accuracy_score(y_test, pred)
accuracy 
```

### rf_clf (Random_Forest)

```python
parameters_rf = {'n_estimators':[50, 70] ,'max_depth':[2,5,10,5.20], 'min_samples_split':[1,5,8,25,35]}
```

```python
grid_rfclf = GridSearchCV(rf_clf, param_grid=parameters_rf, scoring='accuracy', cv=5)
grid_rfclf.fit(X_train, y_train)
```

```python
pred_rfclf = best_rfclf.predict(X_test)
pred_rfclf
```

```python
accuracy_rfclf = accuracy_score(y_test, pred_rfclf)
accuracy_rfclf
```
